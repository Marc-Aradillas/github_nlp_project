# acquire function

```python
import acquire as a
import re
import time
import json

# Define the URL template without the page number
url_template = "https://github.com/search?q=robotics+stars%3A%3E200&type=repositories&p={page}"

# Set the initial page to 1
page = 0
pages = 68
rate_limit_reset = 0  # Initialize to 0 to ensure the first request is made

repositories = []

while page <= pages:
    page += 1

    # Build the URL for the current page
    url = url_template.format(page=page)

    # Check if it's time to sleep to avoid rate limiting
    current_time = time.time()
    if current_time < rate_limit_reset:
        time_to_sleep = rate_limit_reset - current_time
        print(f"Rate limit exceeded. Sleeping for {time_to_sleep} seconds.")
        time.sleep(time_to_sleep)

    # Make the request
    repos = a.github_api_request(url)

    # Check if there are results in the payload
    if 'payload' in repos and 'results' in repos['payload']:
        results = []
        for result in repos['payload']['results']:
            repo_name = result['hl_name']
            repo_name = re.sub(r"<\/?em>", "", repo_name)
            results.append(repo_name)
            print(repo_name)

        # Extend the repositories list only if there are new results
        if results:
            repositories.extend(results)
    else:
        print(f"No results found on page {page}\n\n")

    # Check if there are more pages
    if page % 5 == 0:
        print(f"\n\nacquired pages: {page}")
        time.sleep(60)

# Now you can save the repository names to a JSON file
with open("repo_names.json", 'w') as file:
    json.dump(repositories, file)
```
